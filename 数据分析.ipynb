```python
import pandas as pd
from matplotlib import pyplot as plt
import seaborn as sns
%matplotlib inline
import pyecharts
import matplotlib as mpl
import os

mpl.rcParams['font.family'] = 'Times New Roman'
font1 = {'family' : 'Times New Roman',
'weight' : 'normal',
'size'   : 20,}
font2 = {'family' : 'Times New Roman',
'weight' : 'normal',
'size'   : 14,}
font3 = {'family' : 'Times New Roman',
'weight' : 'normal',
'size'   : 25,}
```

```python
def read_data(name):
    return pd.read_csv(name, sep=r'\t', header=1)
```

```python
def plot_3(data):
    time = range(data.shape[0])
    fig = plt.figure(figsize=[26,8])
    ax1 = fig.add_subplot(131)
    ax1.plot(time, data.iloc[:,2],label='ax(g)')
    ax1.plot(time, data.iloc[:,3],label='ay(g)')
    ax1.plot(time, data.iloc[:,4],label='az(g)')
    ax1.set_title('Acceleration curve', font1)
    ax1.set_xlabel('Time/s',  font2)
    ax1.set_ylabel('acceleration/g',  font2)
    ax1.legend()
    ax1.grid(True)
    
    ax2 = fig.add_subplot(132)
    ax2.plot(time,data.iloc[:,5],label='wx(deg/s)')
    ax2.plot(time,data.iloc[:,6],label='wy(deg/s)')
    ax2.plot(time,data.iloc[:,7],label='wz(deg/s)')
    ax2.set_title('Angular velocity curve',  font1)
    ax2.set_xlabel('Time/s', font2)
    ax2.set_ylabel('Angular velocity/deg/s', font2)
    ax2.legend()
    ax2.grid(True)
    
    ax3 = fig.add_subplot(133)
    ax3.plot(time,data.iloc[:,8], label='AngleX(deg)')
    ax3.plot(time,data.iloc[:,9], label='AngleY(deg)')
    ax3.plot(time,data.iloc[:,10],label='AngleZ(deg)')
    ax3.set_title('Angular curve',  font1)
    ax3.set_xlabel('Time/s', font2)
    ax3.set_ylabel('Angular /deg', font2)
    ax3.legend()
    ax3.grid(True)
    
    fig.show()
```

```python
def plot_data(name):
    data = read_data(name)
    plot_3(data)
```

### x轴旋转一周测试

```python
plot_data('数据4.txt')
```

### y轴旋转一周测试

```python
plot_data('Data_y.txt')
```

### z轴旋转一周测试

```python
plot_data('Data_z.txt')
```
